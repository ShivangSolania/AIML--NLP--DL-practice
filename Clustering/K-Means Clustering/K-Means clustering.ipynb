```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_means = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_means==0, 0], X[y_means==0, 1], s=50, c='red', label = 'C1')
plt.scatter(X[y_means==1, 0], X[y_means==1, 1], s=50, c='blue', label = 'C2')
plt.scatter(X[y_means==2, 0], X[y_means==2, 1], s=50, c='green', label = 'C3')
plt.scatter(X[y_means==3, 0], X[y_means==3, 1], s=50, c='cyan', label = 'C4')
plt.scatter(X[y_means==4, 0], X[y_means==4, 1], s=50, c='magenta', label = 'C5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='black', label = 'Centers')
plt.legend()
plt.show()
```
